```python
# URL da API - Pessoas de 14 anos ou mais de idade que realizam outras formas de trabalho, por sexo e condição de domicílio.
url_r = 'https://servicodados.ibge.gov.br/api/v3/agregados/7070/variaveis?localidades=N7'

# Fazendo a requisição à API
requisicao_r = requests.get(url_r)
```

```python
def check_api_status(url):
    try:
        response = requests.get(url)
        if response.status_code != 200:
            notify("Erro API", f"Status Code: {response.status_code}", "Chegar API", 10)
        else:
            print("API funcionando.")
    except requests.exceptions.RequestException as e:
        notify("Erro API", f"Exception: {e}", "Checar API", 10)

def notify(title, message, app_name, timeout):
    notification.notify(
        title=title,
        message=message,
        app_name=app_name,
        timeout=timeout
    )
```

```python
# URL da API que você deseja verificar
api_url = 'https://servicodados.ibge.gov.br/api/v3/agregados/7070/variaveis?localidades=N7'
check_api_status(api_url)
```

```python
data_r = requisicao_r.json()
```

```python
pprint.pprint(data_r)
```

```python
def process_json(data_r):
    rows = []
    for item in data_r:
        variavel = item['variavel']
        unidade = item['unidade']
        for resultado in item['resultados']:
            for serie in resultado['series']:
                localidade = serie['localidade']['nome']
                nivel = serie['localidade']['nivel']['nome']
                for ano, valor in serie['serie'].items():
                    rows.append({
                        'variavel': variavel,
                        'nivel': nivel,
                        'localidade': localidade,
                        'ano': ano,
                        'valor': valor,
                        'unidade': unidade
                        
                    })
    return pd.DataFrame(rows)
```

```python
# Transformar dados JSON em DataFrame
df_regiao = process_json(data_r)

# Exibir o DataFrame
df_regiao.head(10)
```
